```python
# Using cobe_functions to pull timeseries for a specific lat, long in the dataset
import cobe_functions
import datetime
import matplotlib.pyplot as plt

dataset_path = 'sst.mon.mean.nc'

time, sst = cobe_functions.get_time_series_at_coordinate_latest_sst(0, 210, dataset_path)

# Using cobe_functions to convert the time in the dataset into a datetime vector
datetime_vector = cobe_functions.load_time(dataset_path)

plt.plot(datetime_vector, sst)
plt.xlabel("Time")
plt.ylabel("SST (Celsius)")
plt.title("SST Replotted")
plt.show()
```
